```python
import os
import sys

import numpy as np

import matplotlib.pyplot as plt

sys.path.append("/home/guangda/repos/labelmaker-mix3d")

from mix3d.datasets.scannet200.scannet200_constants import CLASS_LABELS_200, SCANNET_COLOR_MAP_200, VALID_CLASS_IDS_200
```

```python
mink_paths = {
    "baseline": "/home/guangda/repos/labelmaker-mix3d/labelmaker_scripts/supp_figure_data/val/scannet200_mink_baseline.json",
    "ALC+ScanNet200 joint (Ours)": "/home/guangda/repos/labelmaker-mix3d/labelmaker_scripts/supp_figure_data/val/scannet200_mink_alc_scannet_joint.json",
    "ALS200 pretrain+ft (Ours)": "/home/guangda/repos/labelmaker-mix3d/labelmaker_scripts/supp_figure_data/val/scannet200_mink_alc_pretrain_ft.json",
}

ptv3_paths = {
    "baseline": "/home/guangda/repos/labelmaker-mix3d/labelmaker_scripts/supp_figure_data/val/scannet200_ptv3.json",
    "ALC+ScanNet200 joint (Ours)": "/home/guangda/repos/labelmaker-mix3d/labelmaker_scripts/supp_figure_data/val/scannet200_ptv3_alc_scannet_joint.json",
    "PPT with ALC (Ours)": "/home/guangda/repos/labelmaker-mix3d/labelmaker_scripts/supp_figure_data/val/scannet200_ptv3_ppt_with_alc.json",
    "ALS200 pretrain+ft (Ours)": "/home/guangda/repos/labelmaker-mix3d/labelmaker_scripts/supp_figure_data/val/scannet200_ptv3_alc_pretrain.json",
}
```

```python
import json
mink_data = {}
for key in mink_paths.keys():
    with open(mink_paths[key]) as f:
        data_item = json.load(f)
    if "mIoU" in data_item.keys():
        data_item.pop("mIoU")
    mink_data[key] = data_item

mink_trials = mink_paths.keys()
```

```python
ptv3_data = {}
for key in ptv3_paths.keys():
    with open(ptv3_paths[key]) as f:
        data_item = json.load(f)
    if "mIoU" in data_item.keys():
        data_item.pop("mIoU")
    ptv3_data[key] = data_item

ptv3_trials = ptv3_paths.keys()
```

```python
import pandas as pd
```

## Class-wise IoU for MinkowskiNet

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.width', 1000)
mink_df = pd.DataFrame(mink_data).fillna(0)
print(mink_df)
```

## Class-wise IoU for Point Transformer V3

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_colwidth', 1024)
pd.set_option('display.width', 1000)
ptv3_df = pd.DataFrame(ptv3_data).fillna(0)
print(ptv3_df)
```

```python
mink_margin = np.min([mink_df["ALC+ScanNet200 joint (Ours)"].to_numpy(), mink_df["ALS200 pretrain+ft (Ours)"].to_numpy()], axis=0) - mink_df["baseline"].to_numpy()

ptv3_margin = np.min([ptv3_df["PPT with ALC (Ours)"].to_numpy(), ptv3_df["ALS200 pretrain+ft (Ours)"].to_numpy()], axis=0) - ptv3_df["baseline"].to_numpy()
```

```python
classes = list(mink_df.index.values)
selected_class = [classes[i] for i in np.argwhere(((ptv3_margin > 0.045) * (mink_margin > 0.045))).reshape(-1)]
```
